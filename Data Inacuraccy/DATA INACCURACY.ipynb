# `Side by side measurement comparison` BP & Master Product

```python
import pandas as pd

# Load the Excel file and sheets into DataFrames
file_path = 'Product_Data_Shopify_EU_Cleaned.xlsx'

# Load the necessary sheets into DataFrames
df_bright_pearl_live = pd.read_excel(file_path, sheet_name='Bright Pearl Live')
df_master_product = pd.read_excel(file_path, sheet_name='Master Product')

# Clean both DataFrames by removing columns with all null values
df_cleaned_bright_pearl = df_bright_pearl_live.dropna(axis=1, how='all')
df_cleaned_master_product = df_master_product.dropna(axis=1, how='all')

# Filter the Bright Pearl Live data where MP Match Status is "MP MATCH"
df_matched_bright_pearl = df_cleaned_bright_pearl[df_cleaned_bright_pearl['MP Match Status'] == 'MP MATCH']

# Merge the two datasets on 'SKU' to align matching products
df_merged = pd.merge(df_matched_bright_pearl, df_cleaned_master_product, on='SKU', suffixes=('_BrightPearl', '_Master'))

# Calculate differences for each dimension
df_merged['Weight_Diff'] = df_merged['Weight_BrightPearl'] - df_merged['Weight_Master']
df_merged['Height_Diff'] = df_merged['Height_BrightPearl'] - df_merged['Height_Master']
df_merged['Width_Diff'] = df_merged['Width_BrightPearl'] - df_merged['Width_Master']
df_merged['Length_Diff'] = df_merged['Length_BrightPearl'] - df_merged['Length_Master']
df_merged['Volume_Diff'] = df_merged['Volume_BrightPearl'] - df_merged['Volume_Master']

# Determine how many products have differences in dimensions
dimension_columns = ['Weight_Diff', 'Height_Diff', 'Width_Diff', 'Length_Diff', 'Volume_Diff']
differences_count = df_merged[dimension_columns].apply(lambda x: x != 0).sum()

# Summary statistics for the differences
differences_summary = df_merged[dimension_columns].describe()

# Display the results
print("Number of matched products with differences in dimensions:")
print(differences_count)

print("\nSummary statistics for dimension differences:")
print(differences_summary)
```

### `MEASUREMENTS MISSING` for all Shopify

```python
import pandas as pd
from openpyxl import load_workbook

# Define file paths for each Excel file
file_paths = {
    'AU': 'Product_Data_Shopify_AU.xlsx',
    'CA': 'Product_Data_Shopify_CA.xlsx',
    'EU': 'Product_Data_Shopify_EU.xlsx',
    'UK': 'Product_Data_Shopify_UK.xlsx',
    'US': 'Product_Data_Shopify_US.xlsx'
}

# Define measurement columns to check
measurement_columns = [
    'Variant Metafield: custom.height [single_line_text_field]',
    'Variant Metafield: custom.width [single_line_text_field]',
    'Variant Metafield: custom.length [single_line_text_field]',
    'Variant Metafield: custom.weight [single_line_text_field]',
    # 'Variant Metafield: custom.volume [single_line_text_field]'
]

# Function to update DataFrame with Measurement Status
def update_measurement_status(df):
    df['Measurement Status'] = df[measurement_columns].apply(
        lambda row: 'Missing Dimensions' if row.isna().any() or (row == 0).any() else 'Dimensions Presented', axis=1)
    # Insert the new column in the third position (index 2)
    cols = df.columns.tolist()
    cols.insert(2, cols.pop(cols.index('Measurement Status')))
    df = df[cols]
    
    return df

# Process each file
for region, path in file_paths.items():
    # Load the Excel file
    df = pd.read_excel(path, sheet_name='Products')  # Adjust sheet_name if different
    
    # Update the DataFrame with the Measurement Status column
    df_updated = update_measurement_status(df)
    
    # Load the workbook and remove the sheet if it exists
    workbook = load_workbook(path)
    if region in workbook.sheetnames:
        del workbook[region]
        workbook.save(path)  # Save after deleting the sheet
    
    # Write the updated DataFrame to a new sheet named after the region variable (e.g., 'AU', 'CA')
    with pd.ExcelWriter(path, mode='a', engine='openpyxl') as writer:
        df_updated.to_excel(writer, sheet_name=region, index=False)

    print(f"File '{path}' has been updated with a new sheet '{region}' containing the Measurement Status column.")

print("Processing complete. All files have been updated with new sheets.")
```

```python
import pandas as pd
from openpyxl import load_workbook

# Define file paths
file_path_bp = 'Shopify_BP_Master_Data.xlsx'
file_path_bundle = 'Bundle.xlsx'

# Define sheet names
sheet_name_bp = 'Bright Pearl Live'
sheet_name_bundle = 'Sheet1'

# Load the Excel files and sheets into DataFrames
df_bp = pd.read_excel(file_path_bp, sheet_name=sheet_name_bp)
df_bundle = pd.read_excel(file_path_bundle, sheet_name=sheet_name_bundle)

# Function to update DataFrame with Bundle Status
def update_bundle_status(df, bundle_skus):
    # Add a new column 'Bundle Status' indicating if SKU is present in 'Bundle' dataset
    df.insert(
        1,  # Insert at the 2nd column position (index 1)
        'Bundle Status',
        df['SKU'].apply(lambda x: 'Bundle' if x in bundle_skus else 'NO bundle')
    )
    return df

# Update the DataFrame with the Bundle Status column
df_bp_updated = update_bundle_status(df_bp, df_bundle['SKU'].values)

# Load the workbook and remove the sheet if it exists to overwrite
workbook = load_workbook(file_path_bp)
if sheet_name_bp in workbook.sheetnames:
    del workbook[sheet_name_bp]
    workbook.save(file_path_bp)  # Save after deleting the sheet

# Write the updated DataFrame back to the 'Bright Pearl Live' sheet
with pd.ExcelWriter(file_path_bp, mode='a', engine='openpyxl') as writer:
    df_bp_updated.to_excel(writer, sheet_name=sheet_name_bp, index=False)

print(f"File '{file_path_bp}' has been updated with the sheet '{sheet_name_bp}' containing the Bundle Status column.")
```

```python
import pandas as pd

# Define file path and sheet names
file_path = 'Shopify_BP_Master_Data.xlsx'
sheet_name_bp = 'Bright Pearl Live'
sheet_name_mp = 'Master Product'

# Load the Excel file and sheets into DataFrames
df_bp = pd.read_excel(file_path, sheet_name=sheet_name_bp)
df_master_product = pd.read_excel(file_path, sheet_name=sheet_name_mp)

# Step 1: Filter rows in 'Bright Pearl Live' where 'MP Match Status' is 'MP MATCH'
df_bp_filtered = df_bp[df_bp['MP Match Status'] == 'MP MATCH']

# Step 2: Merge the filtered 'Bright Pearl Live' DataFrame with 'Master Product' DataFrame on 'SKU'
df_merged = pd.merge(
    df_bp_filtered,
    df_master_product[['SKU', 'Barcode', 'EAN', 'UPC']],
    on='SKU',
    suffixes=('_BPLive', '_MasterProduct')
)

# Step 3: Normalize EAN, Barcode, and UPC columns to strings and remove trailing '.0'
df_merged['Barcode_BPLive'] = df_merged['Barcode_BPLive'].apply(lambda x: str(x).rstrip('.0'))
df_merged['Barcode_MasterProduct'] = df_merged['Barcode_MasterProduct'].apply(lambda x: str(x).rstrip('.0'))
df_merged['EAN_BPLive'] = df_merged['EAN_BPLive'].apply(lambda x: str(x).rstrip('.0'))
df_merged['EAN_MasterProduct'] = df_merged['EAN_MasterProduct'].apply(lambda x: str(x).rstrip('.0'))
df_merged['UPC_BPLive'] = df_merged['UPC_BPLive'].apply(lambda x: str(x).rstrip('.0'))
df_merged['UPC_MasterProduct'] = df_merged['UPC_MasterProduct'].apply(lambda x: str(x).rstrip('.0'))

# Step 4: Check for exact matches in Barcode, EAN, and UPC
df_exact_match = df_merged[
    (df_merged['Barcode_BPLive'] == df_merged['Barcode_MasterProduct']) &
    (df_merged['EAN_BPLive'] == df_merged['EAN_MasterProduct']) &
    (df_merged['UPC_BPLive'] == df_merged['UPC_MasterProduct'])
]

df_not_exact_match = df_merged[
    (df_merged['Barcode_BPLive'] != df_merged['Barcode_MasterProduct']) |
    (df_merged['EAN_BPLive'] != df_merged['EAN_MasterProduct']) |
    (df_merged['UPC_BPLive'] != df_merged['UPC_MasterProduct'])
]

# Display the results
print("Products with exact matches for Barcode, EAN, and UPC between 'Bright Pearl Live' and 'Master Product':")
df_exact = df_exact_match[['SKU', 'Barcode_BPLive', 'Barcode_MasterProduct', 'EAN_BPLive', 'EAN_MasterProduct', 'UPC_BPLive', 'UPC_MasterProduct']]

print("\nProducts with mismatches for Barcode, EAN, or UPC between 'Bright Pearl Live' and 'Master Product':")
df_not_exact = df_not_exact_match[['SKU', 'Barcode_BPLive', 'Barcode_MasterProduct', 'EAN_BPLive', 'EAN_MasterProduct', 'UPC_BPLive', 'UPC_MasterProduct']]
```

# `Let's create a monthly automatic check for the missing values from the same file and make a table`

Excluding bundles for barcodes.
Excluding bundles for dimensions.

```python
import pandas as pd

# Load the Excel file and specify the required sheets
file_path = 'Dimensions_Barcodes_to_cover_with_Master_Data.xlsx'
sheets = ['AU', 'CA', 'EU', 'UK', 'US']

# Initialize an empty dictionary to store results for each sheet
results = {}

# Loop through each sheet and perform the operations
for sheet in sheets:
    # Read the data from the current sheet
    df = pd.read_excel(file_path, sheet_name=sheet)
    
    # Filter out rows with empty 'Variant SKU' and where 'Variant Metafield: custom_fields.bundle_sku [single_line_text_field]' is 'Yes'
    df_filtered = df[(df['Variant SKU'].notna()) & 
                     (df['Variant Metafield: custom_fields.bundle_sku [single_line_text_field]'] != 'Yes')]
    
    # Count how many Variant SKUs are present (denominator)
    variant_sku_count = df_filtered['Variant SKU'].count()
    
    # Filter rows with missing 'Variant Barcode'
    missing_barcodes_df = df_filtered[df_filtered['Variant Barcode'].isna()]
    
    # Count the number of missing barcodes (nominator)
    missing_barcodes_count = missing_barcodes_df['Variant SKU'].count()
    
    # Calculate the missing barcode percentage
    if variant_sku_count > 0:
        missing_barcode_percentage = (missing_barcodes_count / variant_sku_count) * 100
    else:
        missing_barcode_percentage = 0
    
    # Filter rows where 'Measurement Status' is 'Missing Dimensions'
    missing_dim_df = df_filtered[df_filtered['Measurement Status'] == 'Missing Dimensions']
    
    # Count the number of missing dimensions (nominator for missing dim percentage)
    missing_dim_count = missing_dim_df['Variant SKU'].count()
    
    # Calculate the missing dimensions percentage
    if variant_sku_count > 0:
        missing_dim_percentage = (missing_dim_count / variant_sku_count) * 100
    else:
        missing_dim_percentage = 0
    
    # Store the results for the current sheet
    results[sheet] = {
        'Total Variant SKUs': variant_sku_count,
        'Missing Barcodes': missing_barcodes_count,
        'Missing Barcode Percentage': missing_barcode_percentage,
        'Missing Dimensions': missing_dim_count,
        'Missing Dim. Percentage': missing_dim_percentage
    }

# Display the results for each sheet
for sheet, result in results.items():
    print(f"Sheet: {sheet}")
    print(f"  Total Variant SKUs (Denominator): {result['Total Variant SKUs']}")
    print(f"  Missing Barcodes (Nominator): {result['Missing Barcodes']}")
    print(f"  Missing Barcode Percentage: {result['Missing Barcode Percentage']:.2f}%")
    print(f"  Missing Dimensions (Nominator): {result['Missing Dimensions']}")
    print(f"  Missing Dim. Percentage: {result['Missing Dim. Percentage']:.2f}%")
    print()
```

### Missing Category Report to Giuseppe & Empty Product Type

```python
import pandas as pd

# Load the Excel file and specify the required sheets
file_path = 'Dimensions_Barcodes_to_cover_with_Master_Data.xlsx'
sheets = ['AU', 'CA', 'EU', 'UK', 'US']

# Create a new Excel writer object
output_file = 'Missing_Product_Type_Filtered_Variant_SKUs.xlsx'
writer = pd.ExcelWriter(output_file, engine='xlsxwriter')

# Initialize a dictionary to store the count of non-empty SKUs with empty Type
empty_type_counts = {}

# Loop through each sheet and perform the required operations
for sheet in sheets:
    # Read the data from the current sheet
    df = pd.read_excel(file_path, sheet_name=sheet)
    
    # Filter out rows with empty 'Variant SKU'
    df_non_empty_sku = df[df['Variant SKU'].notna()]
    
    # Denominator: Count of non-empty Variant SKU
    non_empty_variant_sku_count = df_non_empty_sku.shape[0]
    
    # Nominator: Count how many non-empty Variant SKUs have empty Type
    empty_type_count = df_non_empty_sku[df_non_empty_sku['Type'].isna()].shape[0]
    
    # Calculate the percentage of empty 'Type' in relation to non-empty Variant SKUs
    if non_empty_variant_sku_count > 0:
        empty_type_percentage = (empty_type_count / non_empty_variant_sku_count) * 100
    else:
        empty_type_percentage = 0
    
    # Store the count for reporting later
    empty_type_counts[sheet] = {
        'Non-empty Variant SKU Count': non_empty_variant_sku_count,
        'Empty Type Count': empty_type_count,
        'Empty Type Percentage': empty_type_percentage
    }
    
    # Filter the rows where Variant SKU is not empty and Type is empty for exporting
    df_filtered = df_non_empty_sku[df_non_empty_sku['Type'].isna()]
    
    # Select only the required columns (including the newly added columns)
    df_filtered = df_filtered[['Variant SKU', 'Title', 'Type', 'Tags', 'Variant Barcode',
                               'Variant Metafield: custom.height [single_line_text_field]',
                               'Variant Metafield: custom.width [single_line_text_field]',
                               'Variant Metafield: custom.length [single_line_text_field]',
                               'Variant Metafield: custom.weight [single_line_text_field]',
                               'Variant Metafield: custom.volume [single_line_text_field]',
                               'Variant Metafield: custom.volumetric_weight [single_line_text_field]']]
    
    # Write the filtered data to the corresponding sheet in the new Excel file
    df_filtered.to_excel(writer, sheet_name=sheet, index=False)

# Close the Excel writer to save the file
writer.close()

# Print the results for each sheet
for sheet, data in empty_type_counts.items():
    print(f"Sheet: {sheet}")
    print(f"  Total Non-empty Variant SKUs: {data['Non-empty Variant SKU Count']}")
    print(f"  Non-empty Variant SKUs with empty Type: {data['Empty Type Count']}")
    print(f"  Percentage of empty Type: {data['Empty Type Percentage']:.2f}%")
    print()

# Overall summary of non-empty Variant SKUs without Type
total_empty_type_skus = sum([data['Empty Type Count'] for data in empty_type_counts.values()])
total_variant_skus = sum([data['Non-empty Variant SKU Count'] for data in empty_type_counts.values()])
if total_variant_skus > 0:
    overall_empty_type_percentage = (total_empty_type_skus / total_variant_skus) * 100
else:
    overall_empty_type_percentage = 0

print(f"Total non-empty Variant SKUs with no Type across all sheets: {total_empty_type_skus}")
print(f"Overall percentage of non-empty Variant SKUs with empty Type: {overall_empty_type_percentage:.2f}%")
```

### Missing Category Report to Giuseppe & Empty Product Type

```python
import pandas as pd

# Load the Excel file and specify the required sheets
file_path = 'Dimensions_Barcodes_to_cover_with_Master_Data.xlsx'
sheets = ['AU', 'CA', 'EU', 'UK', 'US']

# Create a new Excel writer object
output_file = 'Filtered_Variant_SKUs_with_Empty_Tags.xlsx'
writer = pd.ExcelWriter(output_file, engine='xlsxwriter')

# Initialize a dictionary to store the count of non-empty SKUs with empty Tags
empty_tags_counts = {}

# Loop through each sheet and perform the required operations
for sheet in sheets:
    # Read the data from the current sheet
    df = pd.read_excel(file_path, sheet_name=sheet)
    
    # Filter out rows with empty 'Variant SKU'
    df_non_empty_sku = df[df['Variant SKU'].notna()]
    
    # Denominator: Count of non-empty Variant SKU
    non_empty_variant_sku_count = df_non_empty_sku.shape[0]
    
    # Nominator: Count how many non-empty Variant SKUs have empty Tags
    empty_tags_count = df_non_empty_sku[df_non_empty_sku['Tags'].isna()].shape[0]
    
    # Calculate the percentage of empty 'Tags' in relation to non-empty Variant SKUs
    if non_empty_variant_sku_count > 0:
        empty_tags_percentage = (empty_tags_count / non_empty_variant_sku_count) * 100
    else:
        empty_tags_percentage = 0
    
    # Store the count for reporting later
    empty_tags_counts[sheet] = {
        'Non-empty Variant SKU Count': non_empty_variant_sku_count,
        'Empty Tags Count': empty_tags_count,
        'Empty Tags Percentage': empty_tags_percentage
    }
    
    # Filter the rows where Variant SKU is not empty and Tags is empty for exporting
    df_filtered = df_non_empty_sku[df_non_empty_sku['Tags'].isna()]
    
    # Select only the required columns
    df_filtered = df_filtered[['Variant SKU', 'Title', 'Type', 'Tags', 'Variant Barcode']]
    
    # Write the filtered data to the corresponding sheet in the new Excel file
    df_filtered.to_excel(writer, sheet_name=sheet, index=False)

# Close the Excel writer to save the file
writer.close()

# Print the results for each sheet
for sheet, data in empty_tags_counts.items():
    print(f"Sheet: {sheet}")
    print(f"  Total Non-empty Variant SKUs: {data['Non-empty Variant SKU Count']}")
    print(f"  Non-empty Variant SKUs with empty Tags: {data['Empty Tags Count']}")
    print(f"  Percentage of empty Tags: {data['Empty Tags Percentage']:.2f}%")
    print()

# Overall summary of non-empty Variant SKUs without Tags
total_empty_tags_skus = sum([data['Empty Tags Count'] for data in empty_tags_counts.values()])
total_variant_skus = sum([data['Non-empty Variant SKU Count'] for data in empty_tags_counts.values()])
if total_variant_skus > 0:
    overall_empty_tags_percentage = (total_empty_tags_skus / total_variant_skus) * 100
else:
    overall_empty_tags_percentage = 0

print(f"Total non-empty Variant SKUs with no Tags across all sheets: {total_empty_tags_skus}")
print(f"Overall percentage of non-empty Variant SKUs with empty Tags: {overall_empty_tags_percentage:.2f}%")
```
